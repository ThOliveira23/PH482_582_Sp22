```python
def plot_scatter(data, x, y, xlabel, ylabel):
    plt.figure(figsize=(10,6))
    plt.scatter(data[x],data[y])
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid()
    plt.savefig(x+'_'+y+'.pdf')
    plt.show()
    
    
def plot_one_relative_scatter(data, metric1, metric2, xlabel, ylabel):
    plt.figure(figsize=(10,6))
    plt.scatter(data[metric1],data[metric2]/data['population'])
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid()
    plt.savefig(metric1+'_'+metric2+'.pdf')
    plt.show()
    

def plot_relative_one_scatter(data, metric1, metric2, xlabel, ylabel):
    plt.figure(figsize=(10,6))
    plt.scatter(data[metric1]/data['population'],data[metric2])
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid()
    plt.savefig(metric1+'_'+metric2+'.pdf')
    plt.show()
    
def plot_both_relative_scatter(data, metric1, metric2, xlabel, ylabel):
    plt.figure(figsize=(10,6))
    plt.scatter(data[metric1]/data['population'],data[metric2]/data['population'])
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid()
    plt.savefig(metric1+'_'+metric2+'.pdf')
    plt.show()
    
```

# Plots

```python
plot_scatter(data, 'total_cases_per_million','total_deaths_per_million','Cases/1M','Deaths/1M')          # cases per million x deaths per million
plot_scatter(data, 'life_expectancy','total_deaths_per_million','life expectancy','Deaths/1M')           # life exp. x deaths/1M
plot_scatter(data, 'median_age','total_deaths_per_million','median age','Deaths/1M')                     # median age. x deaths/1M
#plot_scatter(data, 'extreme_poverty','total_deaths_per_million','extreme_poverty','Deaths/1M')           # extreme poverty x deaths/1M

plot_one_relative_scatter(data,'human_development_index','people_fully_vaccinated','HDI','vac.ratio')    # HDI x vac. ratio
plot_one_relative_scatter(data,'gdp_per_capita','people_fully_vaccinated','GDP','vac.ratio')             # HDI x vac. ratio
plot_one_relative_scatter(data,'gdp_per_capita','total_tests','GDP','no tests/population')               # HDI x vac. ratio
#plot_one_relative_scatter(data,'life_expectancy','total_deaths','life expectancy','vac.ratio')          # life exp. x deaths/population
plot_both_relative_scatter(data,'gdp_per_capita','total_tests','gdp/population','tests/population')       # 
```
